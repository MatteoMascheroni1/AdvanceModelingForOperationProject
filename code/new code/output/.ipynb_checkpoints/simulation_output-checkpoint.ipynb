```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("./dataframe.csv")
for i in range(1,6):
    plt.plot(df["UL_in_buffer_"+str(i)], label = i)
plt.legend()
plt.show()
```

```python
df.groupby(["Tugger N"]).agg(Idle_1 = pd.NamedAgg(column="Idle_time_1", aggfunc=lambda x: max(x)/60),
                             Idle_2 = pd.NamedAgg(column="Idle_time_2", aggfunc=lambda x: max(x)/60),
                             Idle_3 = pd.NamedAgg(column="Idle_time_3", aggfunc=lambda x: max(x)/60),
                             Idle_4 = pd.NamedAgg(column="Idle_time_4", aggfunc=lambda x: max(x)/60),
                             Idle_5 = pd.NamedAgg(column="Idle_time_5", aggfunc=lambda x: max(x)/60),
                            )
```

```python
df.groupby(["Tugger N", "Buffer"]).agg(Idle_1=pd.NamedAgg(column="Idle_time_1", aggfunc=lambda x: max(x)/60),
                                          Idle_2=pd.NamedAgg(column="Idle_time_2", aggfunc=lambda x: max(x)/60),
                                          Idle_3=pd.NamedAgg(column="Idle_time_3", aggfunc=lambda x: max(x)/60),
                                          Idle_4=pd.NamedAgg(column="Idle_time_4", aggfunc=lambda x: max(x)/60),
                                          Idle_5=pd.NamedAgg(column="Idle_time_5", aggfunc=lambda x: max(x)/60),
                                       Sat_1 = pd.NamedAgg(column="Saturation_1", aggfunc=lambda x: sum(x)/len(x)*100),
                                       Sat_2 = pd.NamedAgg(column="Saturation_2", aggfunc=lambda x: sum(x)/len(x)*100),
                            )
```
